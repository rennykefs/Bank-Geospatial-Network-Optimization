# Phase 1: Data Ingestion & Spatial Engineering - Interactive Data Exploration (Windows)
This Jupyter Notebook serves as an interactive environment to explore the data ingested into your PostGIS database during Phase 1. It allows for quick visualization, sanity checks, and initial insights into the spatial datasets.

```python
# 1. Setup and Database Connectionimport geopandas as gpdfrom sqlalchemy import create_engineimport matplotlib.pyplot as pltimport os
# Database connection details (ensure these match your setup)DB_USER = "geospatial"DB_PASSWORD = "geospatial_password" # IMPORTANT: Use your actual passwordDB_HOST = "localhost"DB_PORT = "5432"DB_NAME = "geospatial_banking_kenya"
db_connection_str = f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"engine = create_engine(db_connection_str)
# Define the base directory for your project.# This should be the root of your 'geospatial_banking_kenya' project folder.# For Windows, ensure this path is correct, e.g., 'C:\\Users\\YourUser\\Documents\\geospatial_banking_kenya'BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath(__file__)))AOI_DIR = os.path.join(BASE_DIR, "phase1_data_ingestion", "aoi_data")
```

## 2. Explore Administrative Boundaries (AOIs)

```python
# Load AOIs for Nairobi, Mombasa, Kisumufig, ax = plt.subplots(1, 1, figsize=(10, 10))
for filename in os.listdir(AOI_DIR):    if filename.endswith("_aoi.geojson"):        city_name = filename.replace("_aoi.geojson", "").replace("_", " ").title()        aoi_path = os.path.join(AOI_DIR, filename)        aoi_gdf = gpd.read_file(aoi_path)        aoi_gdf.plot(ax=ax, edgecolor="black", facecolor="none", label=city_name)
ax.set_title("Areas of Interest (AOIs) for Kenyan Cities")ax.legend()plt.show()
```

## 3. Explore OpenStreetMap (OSM) Data
Load and visualize road networks and Points of Interest (POIs) for each city.

```python
cities_lower = ["nairobi", "mombasa", "kisumu"]
for city_name_lower in cities_lower:    print(f"
--- Exploring OSM data for {city_name_lower.title()} ---")    try:        # Load road network edges        edges_table = f"{city_name_lower}_osm_edges"        edges_gdf = gpd.read_postgis(f"SELECT * FROM {edges_table}", engine, geom_col="geometry")        print(f"Loaded {len(edges_gdf)} road segments for {city_name_lower.title()}.")
        # Load POIs        pois_table = f"{city_name_lower}_osm_pois"        pois_gdf = gpd.read_postgis(f"SELECT * FROM {pois_table}", engine, geom_col="geometry")        print(f"Loaded {len(pois_gdf)} POIs for {city_name_lower.title()}.")
        # Plotting        fig, ax = plt.subplots(1, 1, figsize=(12, 12))        edges_gdf.plot(ax=ax, color="gray", linewidth=0.5, alpha=0.7)        pois_gdf.plot(ax=ax, color="red", markersize=5, label="POIs")        ax.set_title(f"OSM Data for {city_name_lower.title()}")        ax.legend()        plt.show()
    except Exception as e:        print(f"Error loading OSM data for {city_name_lower.title()}: {e}")
```

## 4. Explore WorldPop Population Data
Load and visualize the gridded population data for each city.

```python
for city_name_lower in cities_lower:    print(f"
--- Exploring WorldPop data for {city_name_lower.title()} ---")    try:        pop_table = f"{city_name_lower}_worldpop_points" # Assuming you converted to points        pop_gdf = gpd.read_postgis(f"SELECT * FROM {pop_table}", engine, geom_col="geometry")        print(f"Loaded {len(pop_gdf)} population points for {city_name_lower.title()}.")
        fig, ax = plt.subplots(1, 1, figsize=(10, 10))        pop_gdf.plot(ax=ax, column="population", cmap="YlOrRd", markersize=5, legend=True)        ax.set_title(f"WorldPop Population for {city_name_lower.title()}")        plt.show()
    except Exception as e:        print(f"Error loading WorldPop data for {city_name_lower.title()}: {e}")
```

## 5. Explore Facebook Mobility Data
Load and visualize the Facebook mobility data for each city.

```python
for city_name_lower in cities_lower:    print(f"
--- Exploring Facebook Mobility data for {city_name_lower.title()} ---")    try:        mobility_table = f"{city_name_lower}_fb_mobility"        mobility_gdf = gpd.read_postgis(f"SELECT * FROM {mobility_table}", engine, geom_col="geometry")        print(f"Loaded {len(mobility_gdf)} mobility records for {city_name_lower.title()}.")
        fig, ax = plt.subplots(1, 1, figsize=(10, 10))        mobility_gdf.plot(ax=ax, markersize=5, alpha=0.5)        ax.set_title(f"Facebook Mobility Data for {city_name_lower.title()}")        plt.show()
    except Exception as e:        print(f"Error loading Facebook Mobility data for {city_name_lower.title()}: {e}")
```
